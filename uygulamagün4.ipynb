```python
!pip install gradio #önce kütüphaneyi indirmek gerek

import gradio as gr

# Hesaplama işlemi fonksiyonu
def matematiksel_islem(sayi1, sayi2, islem):
    if islem == 'Toplama':
        return sayi1 + sayi2
    elif islem == 'Çıkarma':
        return sayi1 - sayi2
    elif islem == 'Çarpma':
        return sayi1 * sayi2
    elif islem == 'Bölme':
        if sayi2 == 0:
            return "Sıfıra bölme hatası!"
        return sayi1 / sayi2

# Arayüz oluşturma
with gr.Blocks() as demo:
    # İki sayı için textbox
    sayi1 = gr.Number(label="Birinci Sayı")
    sayi2 = gr.Number(label="İkinci Sayı")

    # İşlem seçimi için dropdown (açılır menü)
    islem = gr.Dropdown(choices=["Toplama", "Çıkarma", "Çarpma", "Bölme"], label="İşlem Seçin")

    # Hesapla butonu ve sonuç gösterimi
    hesapla_butonu = gr.Button("Hesapla")
    sonuc = gr.Textbox(label="Sonuç")

    # Butona tıklayınca fonksiyon çalıştırılır
    hesapla_butonu.click(matematiksel_islem, inputs=[sayi1, sayi2, islem], outputs=sonuc)

# Gradio uygulamasını başlat
demo.launch()
```
